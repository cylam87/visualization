```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
#adding a legend 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'best')
fig.show()
```

```python
#customizing the legend 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'lower right', 
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True)
fig.show()
```

```python
#moving the legend out of the plot area 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = "upper left", #loc argument is still necessary despite using bbox_to_anchor argument, i.e. bbox_to_anchor does not overwrite loc entirely 
                                #in the presence of bbox_to_anchor loc specifies which edge/corner of the legend is attached to the coordinate specified by bbox_to_anchor
                                #e.g. if loc = "bottom left" and bbox_to_anchor = (1, 1), the bottom left corner legend is attached to the top right corner of the plot
          bbox_to_anchor = (1, 1)) #this is like graph coordinates for a positive-values-only graph; 0,0 is the bottom left corner and 1,1 is the top right corner
fig.show()
```

```python
#adding text annotations: annotations attached to a data point
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
#customizing text annotations
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.text(10, 95, "This value is important!", 
        ha = 'center', #text alignment 
        color = 'red', #text color
        size = 20) #font size 
fig.show()
```

```python
#adding text annotations: annotations attached to a fixed point on fig/ax 
fig,ax = plt.subplots() 
ax.axis([0,10,0,10]) #set ax to range from 0 to 10 on both x and y axes 
ax.text(1, 5, ". Data: (1, 5)", transform = ax.transData) #this is the same as the previous plot - transform = transData attaches the annotation to a data point at x = 1, y = 5. 
                                                            #if not specified, transData is default
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform = ax.transAxes) #transform = transAxes attaches the annotation to the axes coordinates rather than data coordinates. note how the annotation 
                                                            #is at 0.5,0.1 of the axes (which ranges from 0,0 in the bottom left corner to 1,1 in top right) rather than 0.5,0.1 of the data
ax.text(0.5, 0.1, ". Figure: (0.5, 0.1)", transform = fig.transFigure) #transform = transFigure attaches the annotation to the *figure* coordinates. note how it's not the same as axes coordinates
```

```python
#adding arrow annotations 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.annotate('This is important!', xy=(10,95), xytext = (20,94), #need coordinates for both arrow & text; arrow will automatically extend to touch the text 
            arrowprops = dict(facecolor='black')) #arrow properties specified in a dictionary 
fig.show()
```

```python
#customizing arrow annotations - see arrow styles here https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.annotate.html 
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc = "lower right")
ax.annotate('This is important!', xy=(10,95), xytext = (20,94), #need coordinates for both arrow & text; arrow will automatically extend to touch the text 
            arrowprops = dict(arrowstyle='fancy', 
                              color = "hotpink")) #arrow properties specified in a dictionary 
fig.show()
```

```python
#customizing axis labels 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) #remove ticks and labels with locator 
ax.xaxis.set_major_formatter(plt.NullFormatter()) #remove labels only with formatter
```

```python
#specifying number of ticks 
fig, ax = plt.subplots(figsize = (5,3)) 
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = "lower right") 
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #optimal values for axis ticks are automatically chosen 
```

```python
#specifying interval of ticks
fig, ax = plt.subplots(figsize = (5,3)) 
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = "lower right") 
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #intervals of 5 on x axis 
fig.show()
```

```python
#rotating axis labels (to avoid overlap between labels) 
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right') 
plt.xticks(rotation=45, ha='right')
fig.show()
```

```python
#Activity: modifying axis title fonts 
font1 = {'family':'serif', 'color':'indigo'}

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')
plt.xlabel("Shiny New X Axis!", fontsize = 18, fontdict = font1) #identical outcome if using ax.set_xlabel instead 
```
