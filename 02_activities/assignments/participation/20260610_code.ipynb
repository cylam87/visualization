```python
#activity: customize the plots above 
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7,3))
ax1.scatter(x1,y1, marker="*", color="green")
ax2.bar(x2,y2, color="orange")
fig.show()

ax1.set_title('Random scatterplot'), ax2.set_title('Random histogram')
ax1.set_xlabel('Random x axis'), ax2.set_xlabel('Random x axis')
ax1.set_ylabel('Random y axis'), ax2.set_ylabel('Random y axis')
ax1.annotate('Important \n value', xy=(15,55), xytext=(30,60), 
             arrowprops = dict(facecolor='black'))
plt.tight_layout()
fig.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], #specify the figures in row 1
                                    ['ax2', 'ax3']], #specify the figures in row 2; note this will make ax3 stretch over both rows; also note that order matters (ax3, ax2 will error)
                                     figsize=(7,4))
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                     ['ax2', 'ax3']], 
                                     figsize=(7,4))
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
plt.show()
```

```python
#note how the axis label for ax1 doesn't appear (it's covered by ax2, which doesn't respect that the axis label makes ax1 larger), and the axis label for ax3 is cut off
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7,4))
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
#can solve the above issue by using constrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7,4), 
                                    layout = "constrained")
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
#here tight layout also works 
#Tight layout adjusts subplots so tick labels, axis labels, and titles don't overlap or leave the figure area; constrained works similarly but on legends and color bars etc instead 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                    ['ax2', 'ax3']], 
                                    figsize=(7,4), 
                                    layout = "tight")
someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel('A Big Label', fontsize=18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

```python
#Calling multiple visualizations on one axis 
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x, y1, color = "indigo")
ax.plot(x, y2, color="red")
```

```python
#adding error bars 

#first, calculate sd
y2_sd = np.std(y2)
y2_sd
```

```python
#here's the plot we want to apply the error bars to 
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")
```

```python
#now add the error bars
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")
ax.errorbar(x, y2, 
            yerr = y2_sd, 
            fmt = "o", #how the data points (not error bars) are represented. could also use e.g. "none", "-", "o-"
            ecolor = "indigo", #color of error bar 
            elinewidth = 4, #width of error bar
            capsize = 6, #size of the cap of the error bar
            capthick = 4) #thickness of the cap 
```

```python
#to show error bars at intervals / not at every single data point
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")
ax.errorbar(x, y2, 
            yerr = y2_sd, 
            errorevery = 2)
```

```python
#Using images from the internet

from PIL import Image #to open images
import requests #to get images from URLs
from io import BytesIO #to store images
```

```python
url = "https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png"
headers = {"User-Agent": "Mozilla/5.0","Accept": "image/*,*/*;q=0.8","Referer": "https://en.wikipedia.org/"} #headers helps identify me as a legit user so the site won't block my attempt to retrieve the image
response = requests.get(url, headers=headers)
image_file = BytesIO(response.content)
image = Image.open(image_file)
image
```

```python
#adding images to the plot 

#make a basic line plot
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color="red")

#overlay a new axis (ax_image) on the figure (ax) to act as a container for our image 
#note the coordinates refer to the FIGURE COORDINATES, not the AXES COORDINATES
ax_image = fig.add_axes([0.1, #x coordinate, i.e. the image is positioned at 10% of way in from the left of the figure
                         0.11, #y coordinate
                         0.15, #image width, i.e. the image is 15% of the size (width) of the entire figure 
                         0.35]) #image height
```

```python
#now use imshow() to add the image. note how the container (ax_image) is stretched to fit the image (as Duffy is very lean)
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color="red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off') #gets rid of the lines of the container 
```

```python
#saving the visualization as a file
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color="red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')


path = '/Users/cylam/Desktop'
filename = '/fig1a.png'

plt.savefig(path+filename, dpi=300)
```
